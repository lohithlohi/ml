```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ft=pd.read_csv("datasets\Food-Truck(Linear Regression Program).csv")
```

```python
ft.head()
```

```python
# X=ft['a']
# Y=ft['b']

X = ft['a'].values
Y = ft['b'].values
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
mean_a=sum(X)/len(X)
mean_b=sum(Y)/len(Y)
print(mean_a)
print(mean_b)
```

```python
sm=0
den=0
for i in range(len(X)):
    sm+=(X[i]-mean_a)*(Y[i]-mean_b)
    den+=(X[i]-mean_a)**2
m=sm/den
print(m)
```

```python
c=mean_b-m*mean_a
```

```python
print(c)
```

```python
line=[m*i+c for i in X]
plt.plot(X,line,label="line fit",color="r")
plt.scatter(X,Y,label="points")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
SSE=0
for i in range(len(Y)):
    SSE+=(Y[i]-line[i])**2
SSE
```

```python
line=np.array(line)
nm=line.mean()

SSR=0
SST=0
for i in line:
    SSR+=(i-nm)**2
SSR


```
